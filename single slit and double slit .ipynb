```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lam = 500e-9       # Wavelength (500 nm)
b = 10e-6          # Slit width (10 microns)
a = 40e-6          # Distance between slit centers (40 microns)
D = 1.0            # Distance to screen (1 meter)
y = np.linspace(-0.05, 0.05, 1000) # Screen coordinates

# Calculation
sin_theta = y / D
beta = (np.pi * b * sin_theta) / lam
alpha = (np.pi * a * sin_theta) / lam

# Single Slit Intensity
I_single = (np.sinc(beta / np.pi))**2

# Double Slit Intensity
I_double = I_single * (np.cos(alpha))**2

# Normalization
I_single /= np.max(I_single)
I_double /= np.max(I_double)

# Plotting
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(y*100, I_single, color='blue')
plt.title('Single Slit Irradiance (b=10m)')
plt.xlabel('Screen Position (cm)')
plt.ylabel('Normalized Irradiance')

plt.subplot(1, 2, 2)
plt.plot(y*100, I_double, color='red')
plt.title('Double Slit Irradiance (a=40m, b=10m)')
plt.xlabel('Screen Position (cm)')

plt.tight_layout()
plt.show()
```
